# Data Cleaning:

```python
delete=pd.DataFrame([["-0.3900","volatile acidity","No","Alot"],["-0.1900","total sulfur dioxide","No","Alot"],["-0.1700","density","No","Alot"]],columns=["Correlation","Column Name","Normalised","Outliers"])
delete
```

```python
df=df.drop(["volatile acidity","total sulfur dioxide","density"],axis=1)
```

```python
df
```

# Removing outliers:

z-score Technique:

```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(df))
z.shape
```

```python
threshold=3
print(np.where(z>3))
```

```python
df_new=df[(z<3).all(axis=1)]
print(df.shape)
print(df_new.shape)
```

# Percentage data loss:

```python
loss_percent=(1599-1475)/1599*100
print(loss_percent)
```

# Model without outliers is df and model with desired outliers is df_new1

```python
x=df_new.iloc[:,:-1]
y=df_new.iloc[:,-1]
```

# Transforming data to remove skewness

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

# Since the outcome has only two variable we willuse binary classification model.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.metrics import accuracy_score
```

# Splitting the dataframe:

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
        print("At random state",i,"The model performs very well")
        print("At random_state:-",i)
        print("Training r2_score is:-",accuracy_score(y_train,pred_train)*100)
        print("Testing r2_score is:-",accuracy_score(y_test,pred_test)*100)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=899)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_test))
```

# Cross validation score for logistic regression

```python
pred_lr = lr.predict(x_test)
from sklearn.model_selection import cross_val_score
lss = accuracy_score(y_test,pred_lr)
for j in range(2,10):
    lsscore = cross_val_score(lr,x,y,cv=j)
    lsc = lsscore.mean()
    print("At cv:-",j)
    print("Cross validation score is:-",lsc*100)
    print("accuracy_score is :-",lss*100)
    print("\n")
```

```python
lsscore_selected = cross_val_score(lr,x,y,cv=4).mean()
print("The cv score is: ", lsscore_selected,"\nThe accuracy score is: ",lss)
```

# AUC ROC Curve:

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(pred_test, y_test)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=10, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=10, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```
